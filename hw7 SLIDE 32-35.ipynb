```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pathlib
from sklearn.ensemble import RandomForestClassifier
import os
```

```python
n = 100  # Pilih 100 sampel pelatihan
np.random.seed(100)  # Menetapkan seed agar hasil acak dapat direproduksi

# Untuk menyimpan fitur dan tabel
fea_lst = []
y_lst = []

# Daftar file dataset 
files = ['1.csv', '2.csv', '3.csv', '4.csv', '5.csv', '6.csv', '7.csv', '8.csv'] 

# Melakukan iterasi pada setiap file dataset
for i, f in enumerate(files):   
    print(i, f)
    
    # Membaca file CSV ke dalam DataFrame
    ry = pd.read_csv(f)
    
    # Mengisi nilai yang hilang menggunakan forward fill dan backward fill
    ry.ffill(inplace=True)
    ry.bfill(inplace=True)
    
    # Mengonversi DataFrame menjadi array numpy
    data = ry.values    
    
    # Memilih n sampel secara acak (atau lebih sedikit jika jumlah data kurang dari n)
    idx = np.random.choice(data.shape[0], min(n, data.shape[0]), replace=False)
    arr = data[idx, :]
    
    # Menyimpan fitur yang dipilih dan label
    fea_lst.append(arr)
    y_lst.append(np.full(arr.shape[0], i))
    
    print(f'ry.shape {ry.shape}, pilih {arr.shape[0]} fitur')

# Menggabungkan seluruh data fitur dan label
fea_name = list(ry.columns)  # Nama fitur dari DataFrame terakhir
x_data = np.vstack(fea_lst)  # Menggabungkan semua array fitur menjadi satu
y_label = np.hstack(y_lst)   # Menggabungkan semua array label menjadi satu

# Mencetak bentuk data dan label
print(y_label)
print(f'x_data shape: {x_data.shape}')
print(f'y_label shape: {y_label.shape}')

# Inisialisasi RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)

# Melatih model dengan data yang ada
clf.fit(x_data, y_label)

# Mencetak akurasi model pada data pelatihan
print(f'Akurasi Model: {clf.score(x_data, y_label)}')

# Menampilkan pentingnya fitur (optional)
fea_dict = {}
for i, fea in enumerate(fea_name):
    fea_dict[fea] = clf.feature_importances_[i]

# Mengurutkan fitur berdasarkan tingkat kepentingannya
fea_import = [(v, k) for k, v in fea_dict.items()]
fea_import = sorted(fea_import, reverse=True)

#Mencetak pentingnya fitur yang telah diurutkan
print(f'Pentingnya Fitur yang Telah Diurutkan: {fea_import}')

#Menyimpan hasil pentingnya fitur ke dalam file CSV
header = "Feature, Importance"
out_path = r'C:\Users\Salasatun Nurul Aini\OneDrive\ドキュメント\SEMESTER 6\AI dan BIG DATA\output hw 7 slide 32-35'  #Menentukan folder untuk menyimpan file output

# Pastikan folder tujuan ada
if not os.path.exists(out_path):
    os.makedirs(out_path)

# Menyimpan file CSV untuk setiap fitur penting
for i, (importance, feature) in enumerate(fea_import):
    file_path = os.path.join(out_path, f'feature_importance_{i}.csv')  #Menggunakan os.path.join untuk memastikan format path yang benar
    with open(file_path, 'w') as f:
        f.write(f"{header}\n")
        f.write(f"{feature}, {importance}\n")

print(f"Hasil fitur penting disimpan dalam {out_path}")
```
